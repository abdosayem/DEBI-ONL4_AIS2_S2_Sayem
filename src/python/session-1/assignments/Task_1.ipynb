```python
#task_1
email ="Amit_ml@gmail.edu"

if email.count("@") != 1 :
    print("error : Invalid email")
else:
    email_parts = email.split("@")
    username = email_parts[0]
    after_at_Symbol = email_parts[1]

    if "." not in after_at_Symbol :
        print("error : Invalid email")
    else:
        # Extract Domain
        domain = after_at_Symbol.split(".")[0]

        # Check domain ending
        if email.endswith(".com"):
            domain_type = "Commercial Domain"
        elif email.endswith(".edu"):
            domain_type = "Educational Domain"
        else:
            domain_type = "Other Domain"

        print("Username:", username)
        print("Domain:", domain)
        print("Domain Type:", domain_type)

        
```

```python
#Task_2
message = "###!!@mocleW EPGTQ!!!6789"

# Extract the core part
core_part = message[6:18:1]
# core_part = "mocleW EPGTQ"

# Split into words
words = core_part.split()
# Reverse the first word
first_word = words[0][::-1]
# Second word
second_word = words[1]
second_word=second_word.replace('E','e ')
# Final decoded message
decoded_message = first_word + second_word

print(decoded_message)
```

```python
#Task_3
message = "&&&**$gnirtS PLIO!!@1234"

# Extract the core part
core_part =message[6:17:1]
# core_part = "gnirtS PLIO"

# Split into words
words = core_part.split()
# Reverse the first word
first_word = words[0][::-1]
second_word = words[1]
# Second word replace
second_word = second_word.replace('I','E')
second_word = second_word.replace('O','U')

decoded_message = first_word + " " + second_word

print(decoded_message)
```
