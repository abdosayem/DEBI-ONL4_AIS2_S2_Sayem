```python
import os
import random

def thanos_project(folder_path, num_files):
    # 1- open the folder (create if not exists)
    if not os.path.exists(folder_path):
        os.makedirs(folder_path)

    # 2- create number of files inside it
    for i in range(1, num_files + 1):
        file_path = os.path.join(folder_path, f"file_{i}.txt")
        with open(file_path, "w") as f:
            f.write(f"This is file number {i}")

    # 3- check the len of files in that folder
    all_files = os.listdir(folder_path)
    print("Number of files before Thanos:", len(all_files))

    # 4- delete random half of this folder
    files_to_delete = random.sample(all_files, len(all_files) // 2)
    for file in files_to_delete:
        os.remove(os.path.join(folder_path, file))

    # 5- check the result
    remaining_files = os.listdir(folder_path)
    print("Number of files after Thanos:", len(remaining_files))
    print("Remaining files:", remaining_files)


# Example usage:
thanos_project("thanos_folder", 10)
```
