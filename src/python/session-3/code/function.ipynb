```python
def factorial_1 (num):
    if num ==0:
       return 1
    if num == 1:
        return 1  
    return num * factorial_1(num-1)
```

```python
factorial_1(5)
```

```python
def is_prime (num):

    for i in range(2 , num):
        if num % i == 0 :
            return False
    return True    
```

```python
is_prime(3)
```
