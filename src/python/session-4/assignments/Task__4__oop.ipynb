```python
class Fibonacci:
    """
    This Class Generates Fibonacci sequence using recursion.

    """

    def __init__(self, n):
        self.n = n   

    def fib(self, n):
        if n <= 1:
            return n
        return self.fib(n-1) + self.fib(n-2)

    def generate(self):
        return [self.fib(i) for i in range(self.n)]


obj = Fibonacci(10)
print(obj.generate())    
```

```python
class RemoveItemFromSet:
    """
    This Class Removes one item from a set.

    """

    def __init__(self, my_set, item):
        self.my_set = my_set
        self.item = item

    def remove_item(self):
        self.my_set.discard(self.item)
        return self.my_set


obj = RemoveItemFromSet({1,2,3,4,5}, 3)
print(obj.remove_item())
```

```python
class RemoveMultipleItems:
    """
    This Class Removes multiple items from a set.

    """

    def __init__(self, my_set, items):
        self.my_set = my_set
        self.items = items

    def remove_items(self):
        return self.my_set - self.items


obj = RemoveMultipleItems({1,2,3,4,5}, {1,4})
print(obj.remove_items())
```

```python
class FindTupleIndex:
    """
    This Class Finds index of an item in a tuple.

    """

    def __init__(self, my_tuple, item):
        self.my_tuple = my_tuple
        self.item = item

    def find_index(self):
        return self.my_tuple.index(self.item)


obj = FindTupleIndex(('apple','banana','orange','grape'), 'orange')
print(obj.find_index())
```

```python
class SentenceToWords:
    """
    This Class Converts sentence to list of words.

    """

    def __init__(self, sentence):
        self.sentence = sentence

    def convert(self):
        return self.sentence.split()


obj = SentenceToWords("Welcome to Amit")
print(obj.convert())
```

```python
class PalindromeChecker:
    """
    This class Checks if a string is palindrome.

    """

    def __init__(self, text):
        self.text = text

    def check(self):
        if self.text == self.text[::-1]:
            return "Palindrome"
        return "Not a Palindrome"


obj = PalindromeChecker("racecar")
print(obj.check())
```

```python
class Factorial:
    """
    This Class Calculates factorial of a number.

    """

    def __init__(self, number):
        self.number = number

    def calculate(self):
        result = 1
        for i in range(1, self.number + 1):
            result *= i
        return result


obj = Factorial(5)
print(obj.calculate())
```

```python
class SquaresList:
    """
    This Class Creates a list of squares from 1 to 10.

    """

    def generate(self):
        return [i**2 for i in range(1, 11)]


obj = SquaresList()
print(obj.generate())
```

```python
class SumEvenNumbers:
    """
    This Class Calculates sum of even numbers in a list.

    """

    def __init__(self, numbers):
        self.numbers = numbers

    def calculate(self):
        return sum(num for num in self.numbers if num % 2 == 0)


obj = SumEvenNumbers([1,2,3,4,5,6])
print(obj.calculate())
```
